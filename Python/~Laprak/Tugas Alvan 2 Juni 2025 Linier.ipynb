```python
# bagaimana hubungan antra karbo dengan kalori
x = data['Carbohydrates']
y = data['Calories']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.xlabel('Karbo')
plt.ylabel('Kalori')
plt.title('Hubungan antara karbo dengan kalori')

plt.show()
```

```python
from scipy import stats
```

```python
m, c, r, p , std_err = stats.linregress(x, y)
```

```python
m
```

```python
c
```

```python
r
```

```python
p
```

```python
std_err
```

```python
def fungsi(x):
  return m*x + c

mymodel = list(map(fungsi, x))
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel, color='red')
plt.xlabel('Karbo')
plt.ylabel('Kalori')
plt.title('Hubungan antara karbo dengan kalori')

plt.show()
```

```python
x = data["Total Fat"]
y = data["Calories"]

m, c, r, p , std_err = stats.linregress(x, y)

def fungsi(x):
  return m*x + c

mymodel = list(map(fungsi, x))

plt.scatter(x, y)
plt.plot(x, mymodel, color='red')
plt.xlabel('Karbo')
plt.ylabel('Kalori')
plt.title('Hubungan antara total fat dengan kalori')

plt.show()
```

```python
x = data["Protein"]
y = data["Calories"]

m, c, r, p , std_err = stats.linregress(x, y)

def fungsi(x):
  return m*x + c

mymodel = list(map(fungsi, x))

plt.scatter(x, y)
plt.plot(x, mymodel, color='red')
plt.xlabel('Karbo')
plt.ylabel('Kalori')
plt.title('Hubungan antara protein dengan kalori')

plt.show()
```
